## Momentum
- 관성을 이용하여 빠른 수렴 또는 local minima 탈출
- 이전의 W들을 기억해서 반영

## Adaptive Learning Rate
- 학습 초반에는 큰 LR, 후반에는 작은 LR로 최적화
  - epoch에서 loss 개선이 없을경우 lr을 조정
  - 일정 epoch 마다 lr 조정
- update가 많았던 w는 lr을 작게  

### AdaGrad
- 각각의 파라미터가 각자의 lr을 가지고있음
- lr을 gradient ^ 2 의 합으로 나눠줌
- lr또한 W와 같은 shape의 Matrix

## Adam = Adaptive LR + Momentum

### Adam
- p1(= 0.9), p2(= 0.999) 2개의 Parameter
- s: momentum, 약 10%의 gradient와 90%의 직전 momentum을 가져옴
- r: learning_rate, 0.1%의 gradient ^ 2, 99.9%의 직전 lr
- s_hat = s / 1 - (p1 ^ t)
- r_hat = r / 1 - (p2 ^ t)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.datasets import fetch_california_housing
california = fetch_california_housing()
```

```python
df = pd.DataFrame(california.data, columns=california.feature_names)
df['TARGET'] = california.target

df.tail()
```

```python
scaler = StandardScaler()
scaler.fit(df.values[:, :-1])
df.values[:, :-1] = scaler.transform(df.values[:, :-1])

df.tail()
```

```python
data = torch.from_numpy(df.values).float()

data.shape
```

```python
x = data[:, :-1]
y = data[:, -1:]

x.shape, y.shape
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
n_epochs = 4000
batch_size = 256
print_interval = 200
```

```python
model = nn.Sequential(
    nn.Linear(x.shape[-1], 6),
    nn.LeakyReLU(),
    nn.Linear(6, 5),
    nn.LeakyReLU(),
    nn.Linear(5, 4),
    nn.LeakyReLU(),
    nn.Linear(4, 3),
    nn.LeakyReLU(),
    nn.Linear(3, y.shape[-1])
)

model
```

```python
optimizer = optim.Adam(model.parameters())
```

```python
for i in range(n_epochs):
    
    indices = torch.randperm(x.shape[0])
    x_ = torch.index_select(x, dim=0, index=indices)
    y_ = torch.index_select(y, dim=0, index=indices)
    
    x_ = x_.split(batch_size, dim=0)
    y_ = y_.split(batch_size, dim=0)
    
    
    y_hat = []
    total_loss = 0
    
    for x_i, y_i in zip(x_, y_):
        
        y_hat_i = model(x_i)
        loss = F.mse_loss(y_hat_i, y_i)

        optimizer.zero_grad()
        loss.backward()

        optimizer.step()
        
        total_loss += float(loss)
        y_hat += [y_hat_i]
        
    total_loss = total_loss / len(x_)
    
    if (i + 1) % print_interval == 0:
        print('Epoch %d: loss=%.4e' % (i + 1, total_loss))
    
y_hat = torch.cat(y_hat)
y = torch.cat(y_, dim=0)
```

```python
y_hat = torch.cat(y_hat, dim=0)
y = torch.cat(y_, dim=0)
```

```python
df = pd.DataFrame(torch.cat([y, y_hat], dim=1).detach().numpy(),
                  columns=['y', 'y_hat'])
sns.pairplot(df, height=5)
plt.show()
```
